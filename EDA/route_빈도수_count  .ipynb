```python
import pandas as pd

chain_df = pd.read_csv('./PJT001_TripChain.csv', encoding='utf-8')
chain_df.head()
```

### 최초승차역- 최종하차역

```python
ride_on_off = pd.DataFrame(columns=["on-off"])
ride_on_off["on-off"] = str(chain_df["'||최초승차역ID||'"]) + '-' + str(chain_df["'||최종하차역ID||'"]) 
```

```python
ride_on_off["on-off"] = chain_df[["'||최초승차역ID||'", "'||최종하차역ID||'"]].apply(lambda x: '_'.join(map(str,x)), axis=1)
chain_df = pd.concat([chain_df, ride_on_off], axis=1)
chain_df.head()
```

```python
from collections import Counter, OrderedDict

counts = Counter(list(ride_on_off["on-off"])).most_common()
on_off_dict = dict()
for i in counts:
    on_off_dict[i[0]] = i[1]
on_off_dict
```

```python
chain_df.head()
```

##### 전체 route 빈도수 count 
(승차1-하차1-승차2-하차2-승차3-하차3-승차4-하차4-승차5-하차5)

```python
ride_on_off["entire_route"] = chain_df[["'||승차역ID1||'", "'||하차역ID1||'", "'||승차역ID2||'", "'||하차역ID2||'", "'||승차역ID3||'", "'||하차역ID3||'", 
                                       "'||승차역ID4||'", "'||하차역ID4||'", "'||승차역ID5||'", "'||하차역ID5||'"]].apply(lambda x: '-'.join(map(str,x)), axis=1)
entire_route = Counter(list(ride_on_off["entire_route"])).most_common()
entire_route_dict = dict(entire_route)
entire_route_dict
```

```python
chain_df = pd.concat([chain_df, ride_on_off["entire_route"]], axis=1)
```

```python
chain_df.head()
```

```python
chain_df['on-off-count'] = chain_df.groupby('on-off')['on-off'].transform('count')
chain_df['entire_route_count'] = chain_df.groupby('entire_route')['entire_route'].transform('count')
```

```python
chain_df.head()
```

#### 최초승차- 최종하차 빈도수 내림차순정렬   

```python
on_off_sort_chain_df = chain_df.sort_values('on-off-count', ascending=False)
on_off_sort_chain_df.head()
```

#### 전체route  빈도수 내림차순정렬   

```python
entire_route_sort_chain_df = chain_df.sort_values('entire_route', ascending=False)
entire_route_sort_chain_df.head()
```
